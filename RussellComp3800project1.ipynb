```python
from google.colab import drive
drive.mount('/content/drive')

# Reload the data into a pandas DataFrame.
file_path = '/content/drive/MyDrive/Health_conditions_among_children_under_age_18__by_selected_characteristics__United_States.csv'
df = pd.read_csv(file_path)

# Define the columns to drop again.
columns_to_drop = ['INDICATOR', 'UNIT', 'UNIT_NUM', 'STUB_NAME_NUM', 'YEAR_NUM', 'AGE_NUM', 'FLAG']

# Drop these columns from the DataFrame df
df.drop(columns=columns_to_drop, axis=1, inplace=True)

# Display the first few rows of the modified DataFrame to verify the changes
display(df.head())
```

```python
display(df.isnull().sum())
```

```python
df['ESTIMATE'].fillna(df['ESTIMATE'].mean(), inplace=True)
df['SE'].fillna(df['SE'].mean(), inplace=True)
display(df.isnull().sum())
```

```python
# Check for duplicate rows
num_duplicates = df.duplicated().sum()
print(f"Number of duplicate rows before removal: {num_duplicates}")

# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Verify that duplicates have been removed
num_duplicates_after = df.duplicated().sum()
print(f"Number of duplicate rows after removal: {num_duplicates_after}")
```

```python
display(df.info())
```

```python
df['YEAR'] = df['YEAR'].astype('category')
display(df.info())
```

```python
print("Unique values in 'PANEL':")
display(df['PANEL'].unique())

print("\nUnique values in 'STUB_NAME':")
display(df['STUB_NAME'].unique())

print("\nUnique values in 'STUB_LABEL':")
display(df['STUB_LABEL'].unique())

print("\nUnique values in 'AGE':")
display(df['AGE'].unique())
```

```python
# Identify numeric columns
numeric_cols = df.select_dtypes(include=['number']).columns
print(f"Numeric columns: {list(numeric_cols)}")

# Identify outliers using IQR
outlier_counts = {}
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outlier_counts[col] = outliers.shape[0]

# Display the count of outliers for each numeric column
print("\nOutlier counts per numeric column (using 1.5 * IQR):")
for col, count in outlier_counts.items():
    print(f"{col}: {count}")
```

```python
# Define outlier thresholds again based on the previous calculation
Q1_estimate = df['ESTIMATE'].quantile(0.25)
Q3_estimate = df['ESTIMATE'].quantile(0.75)
IQR_estimate = Q3_estimate - Q1_estimate
lower_bound_estimate = Q1_estimate - 1.5 * IQR_estimate
upper_bound_estimate = Q3_estimate + 1.5 * IQR_estimate

# Remove rows where 'ESTIMATE' is an outlier
df_cleaned = df[(df['ESTIMATE'] >= lower_bound_estimate) & (df['ESTIMATE'] <= upper_bound_estimate)].copy()

print(f"Original DataFrame shape: {df.shape}")
print(f"DataFrame shape after removing 'ESTIMATE' outliers: {df_cleaned.shape}")

# Display the first few rows of the cleaned DataFrame
display(df_cleaned.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter for total population (Under 18 years)
df_total = df_cleaned[df_cleaned['STUB_NAME'] == 'Total'].copy()

# Group by Year and Panel and calculate the mean Estimate
df_trend = df_total.groupby(['YEAR', 'PANEL'])['ESTIMATE'].mean().reset_index()

# Create the line chart
plt.figure(figsize=(12, 8))
sns.lineplot(data=df_trend, x='YEAR', y='ESTIMATE', hue='PANEL', marker='o')

# Add titles and labels
plt.title('Prevalence Rates of Health Conditions Among U.S. Children Under 18 Over Time (Total Population)')
plt.xlabel('Year')
plt.ylabel('Mean Estimate (%)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

# Display the chart
plt.show()
```

```python
# Filter for the 'Age' demographic category
df_age = df_cleaned[df_cleaned['STUB_NAME'] == 'Age'].copy()

# Group by AGE and PANEL and calculate the mean ESTIMATE
df_age_grouped = df_age.groupby(['AGE', 'PANEL'])['ESTIMATE'].mean().reset_index()

# Create a grouped bar chart for Age
plt.figure(figsize=(14, 8))
sns.barplot(data=df_age_grouped, x='AGE', y='ESTIMATE', hue='PANEL')

# Add titles and labels
plt.title('Prevalence of Health Conditions by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Mean Estimate (%)')
plt.xticks(rotation=45)
plt.legend(title='Health Condition', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Display the chart
plt.show()
```
